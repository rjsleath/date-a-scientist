```python
df.replace('', np.nan, inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
essay_cols = ["essay0","essay1","essay2","essay3","essay4","essay5","essay6","essay7","essay8","essay9"]
all_essays = df[essay_cols].replace(np.nan, '', regex=True)
all_essays = all_essays[essay_cols].apply(lambda x: ' '.join(x), axis=1)
df["essay_len"] = all_essays.apply(lambda x: len(x))
df.essay_len.head()
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
df.replace(np.nan, '0',inplace=True)
normalised_data = df[['age','essay_len', 'body_code', 'income', 'drinks_code', 'smokes_code']].astype(float)


x = normalised_data.values
min_max_scaler = sklearn.preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)


normalised_data = pd.DataFrame(x_scaled, columns=normalised_data.columns)

normalised_data.head(15)
```

```python

normalised_data.head()
```

```python
#Kmeans
columns = ['age','essay_len', 'income', 'drinks_code', 'smokes_code'] 

X = pd.DataFrame(normalised_data, columns=columns) # load the dataset as a pandas data frame
y = df['body_type'] # define the target variable as y


x_train, x_test, y_train, y_test = train_test_split(X,y,train_size= 0.8, test_size =0.2, random_state=6 )
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(x_train, y_train)
guesses = classifier.predict(x_test)
```

```python
print(guesses)
```

```python
print(classifier.score(x_test, y_test))
```

```python
accuracies=[]
k_list=range(1,101)
for k in range(1, 101):
    classifier = KNeighborsClassifier(n_neighbors = k)
    classifier.fit(x_train, y_train)
    classifier.score(x_test, y_test)
    accuracies.append(classifier.score(x_test, y_test))
    
```

```python
plt.plot(k_list,accuracies)
plt.title("K Value vs Accuracy")
plt.xlabel("K Value")
plt.ylabel("Accuracy")
plt.show()
```

```python
plt.plot(k_list,accuracies)
plt.xlim(80,100)
plt.title("K Value vs Accuracy: Upper Range")
plt.xlabel("K Value")
plt.ylabel("Accuracy")
plt.show()
```

```python
#SVC 1
classifier= SVC(kernel = 'linear')
classifier.fit(x_train, y_train)
print(classifier.score(x_test, y_test))
```

```python
#SVC 2
classifier= SVC(kernel = 'poly')
classifier.fit(x_train, y_train)
print(classifier.score(x_test, y_test))
```

```python
#multilinear regression
columns = ['essay_len','income', 'drinks_code', 'smokes_code'] 

X = pd.DataFrame(normalised_data, columns=columns) # load the dataset as a pandas data frame
y = df['age'] # define the target variable as y


x_train, x_test, y_train, y_test = train_test_split(X,y,train_size= 0.8, test_size =0.2, random_state=6 )
```

```python
mlr = LinearRegression()
mlr.fit(x_train, y_train)
y_predict = mlr.predict(x_test)
```

```python
plt.scatter(y_test, y_predict)
plt.xlabel('y_test')
plt.ylabel('y_predict')
plt.title('Age Actual vs Predicted Values')
plt.show()
```

```python
print(mlr.coef_)
#The first column (essay_len) has the biggest coefficient by far, so it has the strongest effect on age. It's positive, so older people write longer essays.
```

```python
print("Train score:")
print(mlr.score(x_train, y_train))

print("Test score:")
print(mlr.score(x_test, y_test))

#Accuracy:  these variables only explain about 2-3% of the variation in age.
```
